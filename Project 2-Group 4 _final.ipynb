### Make the essential changes to (Revenues and Total cost) columns to change their type

```python
df['Revenues'] = df["Revenues"].str.replace("$","")
df['Revenues'] = df["Revenues"].str.replace(",","")
df['Total Cost'] = df["Total Cost"].str.replace("$","")
df['Total Cost'] = df["Total Cost"].str.replace(",","")
```

```python
df
```

### Check if there is any null values

```python
df.isnull().sum()
```

### Drop unnecessary columns

```python
df.drop(columns=['Code' , 'Volume'] , inplace=True)
```

```python
df
```

### Drop the row that contain null value

```python
df.drop(df.index[(df["Summary Profits"].isnull())] , axis=0 , inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df
```

###  Change the type of (Reveneus and  Total cost) coulmn to  a float64 

```python
df['Revenues'] = df['Revenues'].astype(np.float64)
df['Total Cost'] = df['Total Cost'].astype(np.float64)
df.dtypes
```

```python
df
```

## Data Visualization

### Make a bar chart for product group and thier average profit percent

```python
profit = df.groupby("Product group").mean("Profit Percent" )
plt.figure(figsize=(15 , 7))
plt.bar(profit.index , profit['Profit Percent'] )
plt.title("Product  Group and AVG Profit% "  , fontsize=18)
plt.xlabel('Product Group' , fontsize=12)
plt.ylabel('Profit Percent', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

### Make a pie chart for producers and thier total revenues

```python
a = df.groupby("Producer").sum("Revenues")
a=a.sort_values(by='Revenues' , ascending=False)
a.head()
revList = [8721055.6 , 4807947.40 ,3610528.16, 2228108.64 , 1894368.00 , 29159990.8 ]
producer = ['North Cascade' , 'J&F' , 'Delta' , 'Bonduelle' , 'Caffe Italiano' , 'The Rest producers']
plt.figure(figsize=(15 , 7))
plt.pie(revList , labels=producer , shadow=True , explode=[0.3 ,0,0,0,0,0] ,autopct='%1.0f%%' )
plt.title("Producers Revenues" , fontsize=18)
plt.show()
 
```

### Make a pie chart to show the revenues of each store type

```python
type_rev=df.groupby('Type').sum()['Revenues']
df['Type'].unique()
```

```python
plt.figure(figsize=(8,6))
plt.pie(type_rev, autopct = '%1.1f%%')
plt.legend(['Convenience stores','Hypermarkets','Supermarkets'],loc=0)
plt.title('Revenues by Store Type', fontsize=18)
plt.show()
```

### Make a bar chart to show the revenues of each product

```python
products = df['Product group'].unique()
revenuse_by_product = df.groupby(df['Product group']).sum()['Revenues']

plt.figure(figsize=(24,14))
plt.bar(products,
        revenuse_by_product, color='#D35400')
plt.title('Revenues by product',fontsize=30)
plt.xlabel('Product', fontsize=20)
plt.ylabel('Revenues in millions ($)',fontsize=20)
plt.show()
```

### Make a scatter chart to show the relation between price per unit and profit percent

```python
plt.figure(figsize=(20,10))
plt.scatter(df['Price per unit'][0:2000],df['Profit Percent'][0:2000])
plt.title('Price per unit Vs. Profit Percent', fontsize=20)
plt.xlabel('Cost per unit ($)', fontsize=20)
plt.ylabel('Profit Percent',fontsize=20)
plt.show()
```

### Make a scatter chart to show the relation between price per unit and profit percent

```python
plt.figure(figsize=(20,10))
cost=df.groupby('Producer').mean().sort_values(by='Total Cost')['Total Cost'].nlargest (5)
plt.plot(cost,  color='black', marker='o', linewidth=3)
plt.title('The five highest-cost producers', fontsize=20)
plt.xlabel('Producers', fontsize=20)
plt.ylabel('Total Cost',fontsize=20)
plt.show() 
```

### Make a histogram chart to show the frequent of profit classification

```python
plt.figure(figsize=(20,10))
plt.hist(df['Summary Profits'],color = "lightblue")
plt.title('The profit classification frequency', fontsize=20)
plt.xlabel('Profit classification', fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.show()
```

### Make a bar chart to show the cost for each product

```python
produ = df['Product group'].unique()
plt.figure(figsize=(24,14))
dst = df.groupby('Product group').mean().sort_values(by='Cost per unit')['Cost per unit']
plt.title('The product cost', fontsize=20)
plt.xlabel('Product', fontsize=20)
plt.ylabel('Cost',fontsize=20)
plt.bar(produ,dst, color='#7eb54e')
```
